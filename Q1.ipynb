```python
# importing the required libraries and functions
import nltk
from nltk.probability import FreqDist
from nltk import word_tokenize
from collections import Counter
nltk.download('punkt')
```

```python
# tokenising the given article to get all the words appeared and also how many times do they appear
fdist = Counter()
with open('/content/drive/MyDrive/test.txt', 'r') as fin:
    fdist.update(word_tokenize(fin.read().lower()))
```

```python
# Checking how mnay words appeared once,twice and more than twice
num1 = 0
num2 = 0
num3 = 0
for key in fdist.keys():
  if fdist[key] == 1:
    num1 += 1
  elif fdist[key] == 2:
    num2 += 1
  else:
    num3 += 1

print(num1 + num2 +num3 == len(fdist.keys()))
print(f'Number of words occuring one time: {num1}')
print(f'Number of words occuring two times: {num2}')
print(f'Number of words occuring more than two times: {num3}')
```

```python
# The words which are describing the article will mostly occuring more than once so I want to make a list of
# words which occur atleast two times in the article. and then we compare the similarity of ingredients with these
# words and rank them accordingly
total_words = list(fdist.keys())
selective_words = []
for key in fdist.keys():
  if fdist[key] > 1:
    selective_words.append(key)
```

```python
ingredients_list = ['Ambrette Seed',
'Apple Cinnamon Granola',
'Arizona Seasoning',
'Americano Coffee',
'Baby Abalone',
'Cadbury Double Decker Chocolate Bar',
'Campari Tomato',
'Celery Soup',
'Chia Meal',
'Crunch Bars',
'Cardamom',
'Giardiniera',
'Hog Maw',
'Mccormick Montreal Steak Seasoning',
'Muesli',
'Mulberry',
'Munch Chocolate',
'Murukku Packet',
'Mango',
'Organic Maize',
'Organic Peruvian Groundcherry',
'Organic Tartar Cream',
'Orange Extract',
'Pickled Cauliflower',
'Pork Chump Chops',
'Pork Lungs',
'Pork Tripe',
'Peanut Butter',
'Smokies Sausage',
'Snickers Spread',
'Strawberry Gelatin',
'Salmon',
'Tomato',
'Tamarind',
'Vegan Carob Chips',
'Vegan Chicken Strips',
'Vegan Chorizo',
'Vegan Marshmallow',
'Vegan Puff Pastry Sheet',
'Vegan Semisweet Chocolate Chips',
'Vegan White Cake',
'Vegetable Stock',
'Vinegar']
```

```python
# Trying to check did any of the word in the ingredients_list appeared in the article
def common_data(list1, list2):
  count = 0
  for x in list1:
    if x in list2:
      count += 1
  return count
```

```python
common_data(selective_words, ingredients_list), common_data(total_words, ingredients_list)
```

```python
# sorting the fdist counter to know in general which words appeared the most in the article
from collections import OrderedDict
 
dict1 = dict(sorted(fdist.items(), key=lambda item: item[1], reverse = True))
print(dict1)
```

```python
# Importing the pretrained language transformer
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
# moving the model to device
import torch
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
# Here comes the main part of the code
# The task at hand is to find the ingredients that are most_similar to the words that appear much in the article
# To measure the similarity first we encode them using the language transformer model imported above
# now to measure the similarity we use the cosine_similarity function imported from sklearn library
similar_words = []
from sklearn.metrics.pairwise import cosine_similarity
for word1 in selective_words:
  embedding1 = model.encode(word1)
  for word2 in ingredients_list:
    embedding2 = model.encode(word2)
    if cosine_similarity([embedding1], [embedding2])[0][0] > 0.5:
      similar_words.append((word1, word2, fdist[word1]*cosine_similarity([embedding1], [embedding2])[0][0]))
# Now to rank them based on their relevance I consider the product of two values
# 1) number of times the word1 occured in the passage
# 2) siilarity of the word2 with word1
```

```python
# Here we get a list of tuples with word1(word in article), word2(ingredient), score
similar_words
```

```python
# we have sorted the above list based on the score of the tuple
ranked_list = sorted(similar_words, key = lambda item: item[2], reverse = True)
```

```python
ranked_list
```

```python
# now we cannot simplt get all the ingrdients because there are ingredients which are relevant to multiple words
# so we might get duplicate words so before appending to the list we have to check if it's a duplicate
ranked_ingredients = []
for item in ranked_list:
  if item[1] not in ranked_ingredients:
    ranked_ingredients.append(item[1])
```

```python
# so these are the final output i.e, the list containing ingredients which are ranked based on their relevance to
# the article
ranked_ingredients
# Finally we can see that vegan products and some organic ingredients are being ranked high which makes sense by
# reading the article. as it is more about vegan and healthy food.
```
